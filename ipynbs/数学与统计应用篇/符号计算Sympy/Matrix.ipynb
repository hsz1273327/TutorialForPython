## 矩阵运算

+ ### 矩阵加法

对应项相加

```python
M = Matrix([[1, 3], [-2, 3]])
N = Matrix([[0, 3], [0, 7]])
```

```python
M+N
```

+ 矩阵乘常数

```python
3*M
```

+ ### 矩阵乘向量 

```python
Matrix([[1, 2, 3], [3, 2, 1]]) * Matrix([0, 1, 1])
```

+ ### 矩阵乘法

```python
M*N
```

+ ### 矩阵的幂

```python
M**2
```

```python
M**-1
```

+ ### 矩阵转置

```python
M = Matrix([[1, 2, 3], [4, 5, 6]])
M
```

```python
M.T
```

+ 求解行列式

```python
M = Matrix([[1, 0, 1], [2, -1, 3], [4, 3, 2]])
M
```

```python
M.det()
```

+ ### `rref()`方法
    通过初等行变换，找出向量组的最大无关组，对矩阵操作，转化为最简形矩阵

```python
M = Matrix([[1, 0, 1, 3], [2, 3, 4, 7], [-1, -3, -3, -4]])
M
```

```python
M.rref()
```

+ ### Nullspace零空间

定义：已知$A$为一个$m\times n$矩阵.的零空间（nullspace），又称核（kernel），是一组由下列公式定义的$n$维向量：

$$
ker(A) = {x \in C^n : Ax = 0}
$$

即线性方程组$AX=0$的所有解x的集合。

```python
M = Matrix([[1, 2, 3, 0, 0], [4, 10, 0, 0, 1]])
M
```

```python
M.nullspace()
```

+ ### Columnspace列空间

列空间是指一个由矩阵的所有的列进行线性组合而形成的空间.sympy中使用`columnspace`方法计算,结果返回张成该空间的基向量

```python
M = Matrix([[1, 1, 2], [2 ,1 , 3], [3 , 1, 4]])
M
```

```python
M.columnspace()
```

+ ### 特征值eigenvals,特征向量eigenvects和对角化diagonalize

```python
M = Matrix([[3, -2,  4, -2], [5,  3, -3, -2], [5, -2,  2, -2], [5, -2, -3,  3]])
M
```

```python
M.eigenvals()
```

```python
M.eigenvects()
```

```python
P, D = M.diagonalize()
P
```

```python
D
```

```python
P*D*P**-1
```

```python
P*D*P**-1 == M
```
